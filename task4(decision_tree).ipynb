### TASK 4- To Explore Decision Tree Algorithm


```python
# Importing libraries in Python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
rl=pd.read_csv('iris.csv')
rl.shape
```

```python
print('Missing values in columns:',rl.isnull().any())
rl.describe()
```

### Crating training set and decision tree classifier


```python
# Separating the target variable 
X = rl[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
Y = rl['Species'].values
# Splitting the dataset into train and test 
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.75, random_state=17)
```

```python
# Defining the decision tree algorithm
df= DecisionTreeClassifier()
df.fit(X_train, y_train)
print('Decision Tree Classifer is Created')
```

## predict the species  

```python
# Comparing Actual vs Predicted
y_pred =df.predict(X_test)
df2=pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
pd.DataFrame(np.c_[y_test, y_pred], columns = ["original","predicted"])
```

```python
from sklearn.metrics import confusion_matrix
mat= confusion_matrix(y_test, y_pred)
print(mat)
df.score(X_test, y_test)
```

### Visualization of Decision tree



```python
fig = plt.figure(1, figsize=(18,18))
_ = tree.plot_tree(df, 
                   feature_names=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],  
                   class_names=['Setosa', 'Versicolor', 'Virginica'],filled=True)
```
